```python
from ema_workbench.util import load_results

experiments, outcomes = load_results('./data/pathways with timing.tar.gz')

policies = set(experiments['policy'])

scores = []
index = []

for policy in policies:

    logical = experiments['policy']==policy

    data = {}

    for key in ['Costs', 'Flood damage (Milj. Euro)', 'Number of casualties']:
        value = outcomes[key][logical]
        data['mean {}'.format(key)] = np.mean(value)
        data['std {}'.format(key)] = np.std(value)
        
    scores.append(data)
    index.append(policy)
    
```

```python
experiments, outcomes = load_results('./data/fourth iteration 5 policies  5000 runs.tar.gz')

logical = experiments['policy']=='Dike 1:500 +0.5m, 1:1000'

data = {}

for key, value in outcomes.items():
    value = value[logical]
#     data.append((key, np.mean(value), np.std(value)))
    data['mean {}'.format(key)] = np.mean(value)
    data['std {}'.format(key)] = np.std(value)

scores.append(data)
index.append('Dike 1:500 +0.5m, 1:1000')
    
score_card = pd.DataFrame(scores, index=index)
```
